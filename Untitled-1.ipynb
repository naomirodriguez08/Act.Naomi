```python
#Librerias
import requests
from bs4 import BeautifulSoup
import re
```

```python
# 1. Obtenemos el HTML de la página (usaremos una de prueba común)
url = "https://elfondoenlinea.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
import pandas as pd


titulos = soup.find_all("p", class_= "titulo-mini-slider")
autores = soup.find_all("p", class_ = "autor-mini-slider")
precio_ul = soup.find_all("ul", class_= "nav fce-precios")

ln =  [n.text for n in titulos] 
ln1 = [n.text for n in autores] 
ln2 = [n.text for n in precio_ul] 




df = pd.DataFrame({"titulos": ln, "autores": ln1, "precio_ul": ln2})

df
```
