```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de ejemplo (puedes cambiar la búsqueda)
url = "https://listado.mercadolibre.com.mx/laptop"

# headers para simular navegador
headers = {
    "User-Agent": "Mozilla/5.0"
}

# solicitar página
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# listas para guardar datos
titulos = []
precios = []

# encontrar productos
productos = soup.find_all("div", class_="ui-search-result__wrapper")


for producto in productos:

    titulo_tag = producto.find("h2") or producto.find("a", class_="poly-component__title")
    if titulo_tag:
        titulos.append(titulo_tag.text.strip())
    else:
        titulos.append("Sin título")

    # precio
    precio = producto.find("span", class_="andes-money-amount__fraction")
    if precio:
        precios.append(precio.text)
    else:
        precios.append(None)

# crear dataframe
df = pd.DataFrame({
    "Titulo": titulos,
    "Precio": precios
})

# guardar en excel
df.to_excel("productos_mercadolibre.xlsx", index=False)

print("Datos guardados en productos_mercadolibre.xlsx")
```
