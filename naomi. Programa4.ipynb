```python
class Grupo:
    def __init__(self):
        self.alumnos = []
        self.promedio_grupal = 0.0
        self.destacados = []

    def leer_datos(self):
        print("Ingrese los datos de 10 alumnos:")
        for i in range(10):
            nombre = input("Nombre del alumno " + str(i + 1) + ": ")
            calificacion = float(input("Calificación de " + nombre + ": "))
            self.alumnos.append([nombre, calificacion])

    def realizar_proceso(self):
        suma = 0
        for alumno in self.alumnos:
            suma = suma + alumno[1]

        self.promedio_grupal = suma / 10

        for alumno in self.alumnos:
            if alumno[1] > 9.5:
                self.destacados.append(alumno)

    def resultados(self):
        print("Promedio grupal:", self.promedio_grupal)

        if len(self.destacados) > 0:
            print("Alumnos con calificación mayor a 9.5:")
            for alumno in self.destacados:
                print("- " + alumno[0] + ": " + str(alumno[1]))
        

grupo = Grupo()
grupo.leer_datos()
grupo.realizar_proceso()
grupo.resultados()


```
